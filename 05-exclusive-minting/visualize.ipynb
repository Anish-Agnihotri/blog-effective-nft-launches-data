```python
import pandas as pd
import matplotlib.ticker as tk
```

```python
df = pd.read_json('tokens-minted.json')
```

```python
ax = df.plot.bar(x='num', y='minted', figsize=(18, 12), title="Number of Stoner Cats minted for each quantity batch")
ax.set_xlabel("Quantity Batches (1 = 1 per tx, 20 = 20 per tx)", fontsize=13)
ax.set_ylabel("Stoner Cats minted", fontsize=13)
```

```python
ax.get_figure().savefig('stoner-cats.png')
```
