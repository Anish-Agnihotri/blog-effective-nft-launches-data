```python
import pandas as pd
import matplotlib.ticker as tk
```

```python
df = pd.read_json('september-blocks.json')
```

```python
ax = df.plot.line(x='block', y='baseFeePerGas', figsize=(18, 12), title="Base Fee Per Gas (Month of September)")
ax.set_xlabel("Block Number", fontsize=13)
ax.set_ylabel("Base Fee Per Gas", fontsize=13)
```

```python
ax.get_figure().savefig('basefee.png')
```
